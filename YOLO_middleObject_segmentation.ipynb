```python
# import cv2
# import numpy as np
# Perform object detection and segmentation
results = modelseg("/content/9.jpg", save=True , show_labels=False,show_boxes=True)

from google.colab.patches import cv2_imshow

# # Load the pre-trained YOLOv8 model
# from ultralytics import YOLO
# modelseg = YOLO("yolov8x-seg.pt")

# # Perform object detection and segmentation
# results = modelseg("input.jpg")

# Extract bounding boxes and center points
bounding_boxes = []
object_centers = []
for detection in results[0].boxes:
    x1, y1, x2, y2 = detection.xyxy[0]
    bounding_boxes.append([x1, y1, x2, y2])
    object_centers.append(((x1 + x2) / 2, (y1 + y2) / 2))

# Calculate the image center point
image_center = (results[0].orig_img.shape[1] // 2, results[0].orig_img.shape[0] // 2)

# Find the middle object
distances = [np.sqrt((cx - image_center[0])**2 + (cy - image_center[1])**2) for cx, cy in object_centers]
middle_object_index = np.argmin(distances)
middle_object_bbox = bounding_boxes[middle_object_index]

# Draw a rectangle around the middle object
image = cv2.imread("/content/9.jpg")
cv2.rectangle(image, (int(middle_object_bbox[0]), int(middle_object_bbox[1])), (int(middle_object_bbox[2]), int(middle_object_bbox[3])), (0, 255, 0), 2)

# Display the output image
cv2_imshow(image)
```
